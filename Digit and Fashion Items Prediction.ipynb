```python
import tensorflow as tf
```

```python
tf.__version__
mnist = tf.keras.datasets.mnist
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()

```

```python
x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

```python
import matplotlib.pyplot as plt
print(x_train[0])
plt.imshow(x_train[0],cmap = plt.cm.binary)
```

```python
model= tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128,activation=tf.nn.softmax))


model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train,y_train,epochs=3)

```

```python
val_loss, val_acc=model.evaluate(x_test,y_test)
print(val_loss,val_acc)
```

```python
from tensorflow import keras
import numpy as np
```

```python
data = keras.datasets.fashion_mnist
```

```python
(train_images, train_labels),(test_images, test_labels)=data.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
train_images=train_images/255
test_images=test_images/255
plt.imshow(train_images[17],cmap=plt.cm.binary)
```

```python
model= keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128,activation='relu'), #rectified Linear Unit
    keras.layers.Dense(10,activation='softmax') #probablity of a certain network thinking its a particular class
    ])

model.compile(optimizer="adam", loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_images,train_labels,epochs=8)# How many times the model is going to see this infornmation
```

```python
test_loss, test_acc= model.evaluate(test_images,test_labels)
```

```python
prediction = model.predict(test_images)
```

```python
class_names[np.argmax(prediction[0])]
```

```python
for i in range(10,20):
    plt.grid(False)
    plt.imshow(test_images[i],cmap=plt.cm.binary)
    print(i)
    plt.xlabel("Actual:"+ class_names[test_labels[i]])
    plt.title("Predictions "+class_names[np.argmax(prediction[i])])
    plt.show()
    
```
